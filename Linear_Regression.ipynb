```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
#using sci-kit learn for data mining and data analysis

from sklearn.linear_model import LinearRegression
```

```python
data=pandas.read_csv("costRevenue.csv")
```

```python
data.describe()
```

```python
#using pandas dataframe 
X= DataFrame(data,columns=['production_budget_usd'])  #production budget 
y= DataFrame(data,columns=['worldwide_gross_usd']) #revenue
```

```python
#using matplotlib to visualize data e.g charts
#showing scatter plot
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget $")
plt.ylabel ("Worldwide Gross $")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression= LinearRegression()
regression.fit(X,y) #running regression numbers have been crunched 
```

regression.coef basically tells about slope coefficient

```python
regression.coef_ #slope co-efficient theta_1 
```

```python
#intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
#calculat predictive value for all x
plt.plot(X, regression.predict(X),color="red",linewidth=4)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget $")
plt.ylabel ("Worldwide Gross $")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
#substiute values in h0(x) formula where x0 is intercept and x1 is co-efficient
regression.score(X,y) #r2 is the amount of variation that is explanied by budget which is 55% in this case
```
